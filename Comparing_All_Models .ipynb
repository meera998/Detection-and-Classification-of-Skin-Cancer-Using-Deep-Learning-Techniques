```python
import pandas as pd 
import numpy as np
import seaborn as sns
from sklearn.metrics import auc


%config InlineBackend.figure_format = 'svg'
plt.rcParams ['figure.figsize'] = (9,7)
sns.set(style='white', context='talk', font_scale=0.8)
```

```python
ResNet18 = pd.read_csv('resnet18_roc_curve.csv')
AlexNet = pd.read_csv('AlexNet_roc_curve.csv')
DenseNet121 = pd.read_csv('densenet121_roc_curve.csv')
```

```python
resnet_fpr = np.array([0,0.512755,1])
resnet_tpr = np.array([0,0.987589, 1 ])

alexnet_fpr= np.array([0, 0.734694, 1])
alexnet_tpr= np.array([0, 0.957151, 1])

densenet_fpr = np.array([0,0.283163,1])
densenet_tpr = np.array([0,0.959811,1])
```

```python
# DenseNet ROC curve
densenet_auc = auc(densenet_fpr, densenet_tpr)
plt.plot(densenet_fpr, densenet_tpr, color='green', lw=2, label='DenseNet(AUC = %0.2f)' % densenet_auc)

# ResNet ROC curve
resnet_auc = auc(resnet_fpr, resnet_tpr)
plt.plot(resnet_fpr, resnet_tpr, color='blue', lw=2, label='ResNet (AUC = %0.2f)' % resnet_auc)

# AlexNet ROC curve
alexnet_auc = auc(alexnet_fpr, alexnet_tpr)
plt.plot(alexnet_fpr, alexnet_tpr, color='red', lw=2, label='AlexNet (AUC = %0.2f)' % alexnet_auc)

# Plotting settings
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for All Models', fontweight='bold')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
